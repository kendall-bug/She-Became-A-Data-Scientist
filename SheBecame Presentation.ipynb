```python
# Data cleansing is a big part of data science - data can be messy and it can affect our results.
# Below are some common strategies that are used for cleaning up text:

# Convert everything to lowercase letters to avoid some words being duplicated.
# For example, Python would treat "you" and "You" as different words if we don't control for this.
text = text.lower()

# We want to remove common words like "that" and "it" that take up space but do not tell us anything specific about 
# the content of the song. Most programming languages have pre-made lists of these words and refer to them as "stop words."
# Create stopword list:
stopwords = set(STOPWORDS)

# You can add your own stopwords if you want to leave certain words out.
# Here, I removed "Ah","ya", "still" and "even" because they don't tell us what the song is about.
stopwords.update(["Ah", "ya", "even"])

```
